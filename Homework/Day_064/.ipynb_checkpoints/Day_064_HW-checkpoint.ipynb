```python
#import required libraries 
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cv2.cvtColor is an OpenCV function to convert images to different color spaces
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detect_faces(f_cascade, colored_img, scaleFactor = 1.1, minNeighbors=5):
 #just making a copy of image passed, so that passed image is not changed 
 img_copy = colored_img.copy()          
 
 #convert the test image to gray image as opencv face detector expects gray images
 gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)          
 
 #let's detect multiscale (some images may be closer to camera than others) images
 faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=minNeighbors);          
 
 #go over list of faces and draw them as rectangles on original colored img
 for (x, y, w, h) in faces:
      cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 20)              
 
 return img_copy, faces
```

```python
#load another image 
test5 = cv2.imread('data/test5.jpg')
```

```python
fig, ax = plt.subplots(1,2,figsize=(10, 7))
#load cascade classifier training file for haarcascade 
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml')
#call our function to detect faces 
faces_detected_img, faces = detect_faces(haar_face_cascade, test5, scaleFactor=1.1, minNeighbors=5) 
#convert image to RGB and show image
ax[0].imshow(convertToRGB(faces_detected_img))
ax[0].set_title(f"num_faces={len(faces)}, {faces}")
#call our function to detect faces 
faces_detected_img, faces = detect_faces(haar_face_cascade, test5, scaleFactor=1.6, minNeighbors=3) 
#convert image to RGB and show image 
ax[1].imshow(convertToRGB(faces_detected_img))
ax[1].set_title(f"num_faces={len(faces)}, {faces}")
```

```python
fig, ax = plt.subplots(1,2,figsize=(10, 7))
#load cascade classifier training file for haarcascade 
lbpcascade = cv2.CascadeClassifier('data/lbpcascade_frontalcatface.xml')
#call our function to detect faces
faces_detected_img,faces = detect_faces(lbpcascade, test5, scaleFactor=1.1, minNeighbors=2) 
#convert image to RGB and show image 
ax[0].imshow(convertToRGB(faces_detected_img))
ax[0].set_title(f"num_faces={len(faces)}, {faces}")
#call our function to detect faces 
faces_detected_img,faces = detect_faces(lbpcascade, test5, scaleFactor=1.6, minNeighbors=1)
#convert image to RGB and show image 
ax[1].imshow(convertToRGB(faces_detected_img))
ax[1].set_title(f"num_faces={len(faces)}, {faces}")
```
